```python
def get_nv(a, b, c):
    ab = b-a
    ac = c-a
    
    n = np.ones([2,3])
    n[0,:] = ab
    n[1,:] = ac
    x1 = n[0,1]*n[1,2] - n[1,1]*n[0,2]
    y1 = n[0,2]*n[1,0] - n[1,2]*n[0,0]
    z1 = n[0,0]*n[1,1] - n[1,0]*n[0,1]
    
    nv = [x1, y1, z1]
    return nv    
```

```python
a = np.array([1,0,0])
b = np.array([0,1,0])
c = np.array([0,0,1])

nv = get_nv(a, b, c)
nv
```
